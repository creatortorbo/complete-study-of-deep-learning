from tensorflow.keras.preprocessing.image import ImageDataGenerator:
This line imports the ImageDataGenerator class from tensorflow.keras.preprocessing.image.
ImageDataGenerator is a tool that allows you to perform data augmentation in real-time by generating
batches of tensor image data with real-time data augmentation. This is commonly used to artificially expand the size 
of your dataset by applying random transformations like rotations, flips, shifts, etc., to the training images.

```python
# here it for the loading data 
imag = image.load_img('IMG_20231030_172417.jpg',target_size=(200,200))
```

```python
plt.imshow(imag)
plt.show()
```

```python
type(imag)
```

Shearing is a transformation that shifts one part of an image in a certain direction while keeping another part fixed

```python
datagen = ImageDataGenerator(rotation_range=30,
                            shear_range=0.2, 
                            zoom_range=0.2,
                            horizontal_flip=True, 
                            width_shift_range=0.2,
                            height_shift_range=0.2
                            )
```

```python
imag = image.img_to_array(imag)
```

```python
imag.shape
```

```python
type(imag)
```

```python
input_batch = imag.reshape(1,200,200,3)
```

```python
i = 0 

for output in datagen.flow(input_batch,batch_size=1,save_to_dir='aug'):
   i = i + 1
   if i == 10: 
        break
```
