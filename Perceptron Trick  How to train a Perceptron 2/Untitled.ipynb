```python
def preceptron(x,y):
    x  = np.insert(x,0,1,axis = 1)  # here in this we insert the 1 in the 0 index 
    weights = np.ones(x.shape[1])
    lr = 0.1 # it learning rate inslizined 

    for i in range(1000):
        j = np.random.randint(0,100)  # here in this selating the random values form 0 to b/w 100
        # form this we just picking the only one form this j form 0 b/w 100
        y_hat = step(np.dot(x[j],weights))  # here in this it comes greater than zero
        # here with the y_hat we predicting that is plced or not placed 
        weights = weights + lr*(y[j]-y_hat)*x[j]
         # here we upated the weight 
    return weights [0] ,weights [1:]
     # here it return the weights and the intercepetes
```
