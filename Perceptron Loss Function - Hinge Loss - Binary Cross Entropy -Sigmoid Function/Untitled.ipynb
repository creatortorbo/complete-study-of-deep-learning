Lr : The learning rate controls how quickly or slowly a model learns

```python
def perceptron(x,y): 
    w1 = w2 = b = 1   # here all this working with gradint desent 
    lr = 0.1

    for j in range(1000): 
          for i in range(x.shape[0]): # here we going into the rows all values of x
                # check condition 
              z = w1*x[i][0] + w2*x[i][1] + b 
              # here z is f(xi) is 
              if z*y[i] < 0:
                  w1 = w1 + lr*y[i]*x[i][0]  # udate role 
                  w2 = w2 + lr*y[i]*x[i][1]
                  b = b + lr*y[i]

    return w1 , w2 ,b 
```

```python
w1,w2,b = perceptron(x,y)
```
